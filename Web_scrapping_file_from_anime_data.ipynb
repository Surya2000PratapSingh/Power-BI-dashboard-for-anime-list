```python
url="https://en.wikipedia.org/wiki/List_of_anime_franchises_by_episode_count"
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r=requests.get(url)
r
```

```python
soup=BeautifulSoup(r.content)
soup
```

```python
table=soup.find_all("table",class_="wikitable sortable")[0]
# print(table)
table_heading=table.find_all("th")
table_heading
```

```python
headings=[t.text.strip() for t in table_heading]
headings
```

```python
df=pd.DataFrame(columns=headings)
df
```

```python
row=table.find_all("tr")
for r in row[1:]:
  k=r.find_all("td")
  individual_row_data=[t.text.strip() for t in k]
  print(individual_row_data)
  length=len(df)
  df.loc[length]=individual_row_data
df
```

```python
df.to_csv(r"List of anime franchises by episode count.csv",index=False)
```
